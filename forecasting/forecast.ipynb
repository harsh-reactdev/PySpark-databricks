```python
toner_usage_schema = StructType([

    StructField("deviceId", StringType(), True),
    StructField("dealerId", StringType(), True),
    StructField("L1DealerId", StringType(), True),
    StructField("L2DealerId", StringType(), True),
    StructField("L3DealerId", StringType(), True),
    StructField("divisionId", StringType(), True),
    StructField("servicingDealerId", StringType(), True),
    StructField("customerId", StringType(), True),

    StructField("modelName", StringType(), True),
    StructField("serialNumber", StringType(), True),

    StructField("timestamp", StringType(), True),
    StructField("lastSuppliesUpdate", StringType(), True),

    StructField("relatedGroupId", StringType(), True),
    StructField("type", StringType(), True),
    StructField("description", StringType(), True),
    StructField("state", StringType(), True),

    StructField("color", StringType(), True),
    StructField("typical", StringType(), True),
    StructField("unit", StringType(), True),
    StructField("capacity", StringType(), True),

    StructField("printCount", IntegerType(), True),
    StructField("tonerNumber", IntegerType(), True),

    StructField("tagIds", StringType(), True),
    StructField("forecastingList", StringType(), True)
])
```

```python
df = spark.read.format('csv').schema(toner_usage_schema).option('header', True).load('/Volumes/workspace/default/forecast/Toner Usage.csv')
display(df)
```

```python
df_base = (
    df
    .select(
        col("deviceId"),
        col("color"),
        to_timestamp(col("timestamp")).alias("timestamp"),
        col("typical").cast("double").alias("typical"),
        col("printCount").cast("int").alias("printCount")
    )
    .dropna(subset=["deviceId", "color", "timestamp", "typical"])
)
# display(df_base)

df_base.orderBy("deviceId", "color", "timestamp").show(10, truncate=False)
df_base.printSchema()
```

```python
w = Window.partitionBy("deviceId", "color").orderBy("timestamp")

# computing previous values
df_deltas = (
    df_base
    .withColumn("prev_timestamp", lag("timestamp").over(w))
    .withColumn("prev_typical", lag("typical").over(w))
)

# computing difference between present and previous values
df_deltas = (
    df_deltas
    .withColumn("daysDiff", datediff(col("timestamp"), col("prev_timestamp")))
    .withColumn("typicalDiff", col("prev_typical") - col("typical"))
)

# verify
df_deltas.select(
    "deviceId", "color", "timestamp",
    "typical", "prev_typical",
    "daysDiff", "typicalDiff"
).show(10, truncate=False)
```

```python
# computing daily depletion in toner levels
df_rate = df_deltas.withColumn(
    "dailyDepletion",
    try_divide(
        col("typicalDiff"), col("daysDiff")
    )
)

df_rate = df_rate.filter(col("dailyDepletion") > 0)

df_rate.select(
    "deviceId", "color", "timestamp",
    "daysDiff", "typicalDiff", "dailyDepletion"
).show(10, truncate=False)

```

```python
df_avg_rate = (
    df_rate
    .groupBy("deviceId", "color")
    .agg(
        round(avg("dailyDepletion"), 2).alias("avgDailyDepletion")
    )
)

display(df_avg_rate)
```

```python
w_last = Window.partitionBy("deviceId", "color").orderBy(col("timestamp").desc())

df_last = (
    df_base
    .withColumn("rn", row_number().over(w_last))
    .filter(col("rn") == 1)
    .select(
        "deviceId",
        "color",
        col("timestamp").alias("lastTimestamp"),
        col("typical").alias("lastTypical")
    )
)

df_last.show(truncate=False)
```

```python
df_forecast = (
    df_last
    .join(df_avg_rate, ["deviceId", "color"], how="inner")
)

df_forecast = df_forecast.withColumn(
    "daysUntilEnd",
    ceil(col("lastTypical") / col("avgDailyDepletion"))
)

df_forecast = df_forecast.withColumn(
    "predictedEndDate",
    date_add(
        col("lastTimestamp"),
        col("daysUntilEnd").cast("int")
    )
)

display(df_forecast)
```

```python
df_predictedEndDate = df_forecast.select(
    "deviceId",
    "color",
    "lastTypical",
    "avgDailyDepletion",
    "daysUntilEnd",
    "predictedEndDate"
)

display(df_predictedEndDate)
```

```python
df_print_rate = df_deltas.withColumn(
    "printsPerTypical",
    round(try_divide(col("printCount"), col("typicalDiff")))
)

df_print_rate = df_print_rate.filter(
    (col("printsPerTypical") > 0) &
    col("printsPerTypical").isNotNull()
)

df_print_rate.select(
    "deviceId",
    "color",
    "printCount",
    "typicalDiff",
    "printsPerTypical"
).show(10, truncate=False)
```

```python
df_avg_print_rate = (
    df_print_rate
    .groupBy("deviceId", "color")
    .agg(
        round(avg("printsPerTypical"), 2).alias("avgPrintsPerTypical")
    )
)
df_avg_print_rate.show(truncate=False)
```

```python
df_print_forecast = (
    df_last
    .join(df_avg_print_rate, ["deviceId", "color"], how="inner")
)

df_print_forecast = df_print_forecast.withColumn(
    "remainingPrints",
    ceil(col("lastTypical") * col("avgPrintsPerTypical"))
)

df_print_forecast.select(
    "deviceId",
    "color",
    "lastTypical",
    "avgPrintsPerTypical",
    "remainingPrints"
).show(truncate=False)
```

```python
df_final = (
    df_forecast
    .join(
        df_print_forecast.select(
            "deviceId",
            "color",
            "remainingPrints"
        ),
        ["deviceId", "color"],
        how="inner"
    )
)

df_final.select(
    "deviceId",
    "color",
    "predictedEndDate",
    "daysUntilEnd",
    "remainingPrints"
).show(truncate=False)

display(df_final)
```
